### Линейное пространство. Основные понятия. Часть 1

__1.__ Исследовать на линейную зависимость:

$$f_{1}(x)=e^{x}, f_{2}(x)=1, f_{3}(x)=x+1, f_{4}(x)=x-e^{x}.$$

#### Решение:
Заметим, что $f_{4}(x) = f_{3}(x) - f_{2}(x) - f_{1}(x)$, то есть вектор $f_{4}(x)$ - линейная комбинация векторов $f_{1}(x)$, $f_{2}(x)$ и $f_{3}(x)$, из чего можно сделать вывод, что $f_{1}(x)=e^{x}, f_{2}(x)=1, f_{3}(x)=x+1, f_{4}(x)=x-e^{x}.$ - линейно зависимы

__2.__ Исследовать на линейную зависимость:
$$f_{1}(x)=2, f_{2}(x)=x, f_{3}(x)=x^{2}, f_{4}(x)=(x+1)^{2}$$

#### Решение:
Заметим что $f_{4}(x) = f_{3}(x) + f_{1}(x) \cdot f_{2}(x) + 1$,  то есть вектор $f_{4}(x)$ - линейная комбинация векторов $f_{1}(x)$, $f_{2}(x)$ и $f_{3}(x)$, из чего можно сделать вывод, что $f_{1}(x)=2, f_{2}(x)=x, f_{3}(x)=x^{2}, f_{4}(x)=(x+1)^{2}$  - линейно зависимы

#### Решение:
Базис линейного пространства $\mathbb{R}^{3}$ образуют векторы $b_{1}=(0, 0, 10)$, $b_{2}=(2, 0, 0)$, $b_{3}=(0, 1, 0)$. 

Тогда 

$$ x = (2,3,5 = (2,0,0) + (0,3,0) + (0,0,5) = 1 \cdot (2,0,0) + 3 \cdot (0,1,0) + 0,5 \cdot (0,0,10) = 1b_{2} + 3b_{3} + 0,5b_{1}  $$

то есть координатами вектора $x$ в базисе являются $1$, $3$, $0.5$.

__5.__ Установить, является ли линейным подпространством:

а) совокупность всех векторов трехмерного пространства, у которых по крайней мере одна из первых двух координат равна нулю;
    
б) все векторы, являющиеся линейными комбинациями данных векторов $\{u_{1}, u_{2}, \ldots, u_{n}\}$.

По определению подмножество $L$ линейного пространства $V$ является его подпространством тогда и только тогда, когда для любых элементов $u, v\in L$ и любого $\alpha \in \mathbb{R}$ выполняются условия:
$$1)\ u+v\in L;$$$$2)\ \alpha\cdot u \in L.$$

Для рассматриваемой в данном пункте совокупности векторов нарушается первое условие. Рассмотрим, например, векторы $u = (0, 1, 1)$ и $v = (1, 0, 1)$. Тогда $u + v = (1, 1, 2) \notin L$. То есть совокупность всех векторов трехмерного пространства, у которых по крайней мере одна из первых двух координат равна нулю не является линейным подпространством.

### Линейное пространство. Основные понятия. Часть 2

__1.__ Найти скалярное произведение векторов $x, y \in \mathbb{R}$:<br>
а) $x=(0,-3, 6),~y=(-4, 7, 9);$<br>
б) $x=(7, -4, 0, 1),~y=(-3, 1, 11, 2).$

#### Решение:
 $$ а) (x, y) = 0 \cdot (- 4) - 3 \cdot 7 + 6 \cdot 9 = 54 - 21 = 33 $$
 $$ б) (x, y) = 7 \cdot (-3) -4 \cdot 1 + 0 \cdot 11 + 1 \cdot 2 = -21 - 4 + 2 = -23 $$

__2.__ Найти нормы векторов $(4, 2, 4)$ и $(12, 3, 4)$ и угол между ними.

#### Решение:
Введем обозначения для векторов: x = (4, 2, 4); y = (12, 3, 4).
Определим манхэттенские нормы векторов ($\ell_{1}$):
$$\left\|x\right\|_{1} = |4| + |2| + |4| = 10 $$
$$\left\|y\right\|_{1} = |12| + |3| + |4| = 19 $$
Определим евклидовы нормы векторов ($\ell_{2}$):
$$ \left\|x\right\|_{2} = \sqrt{ 4^2 + 2^2 + 4^2 } = 6 $$
$$ \left\|y\right\|_{2} = \sqrt{ 12^2 + 3^2 + 4^2 } = 13 $$
Определим угол между данными векторами:
$$ cos\varphi = \frac{(x, y)}{\left\|x\right\|_{2} \cdot \left\|y\right\|_{2}} = \frac{4 \cdot 12 + 2 \cdot 3 + 4 \cdot 4}{6 \cdot 13} = \frac{70}{78} \approx 0.897 $$


__3.__ Будет ли линейное пространство евклидовым, если за скалярное произведение принять:<br>
а) произведение длин векторов;<br>
б) утроенное обычное скалярное произведение векторов?<br>

#### Решение:
а) В таком случае $ (x, y ) = \left \|x\right \| \cdot \left \|y\right \|$. Согласно определению, линейное пространство над полем вещественных чисел называется _евклидовым пространством_, если на нем введено правило, согласно которому каждой упорядоченной паре векторов $x$ и $y$ поставлено в соответствие вещественное число, называемое _скалярным произведением_ и обозначаемое символом $( x, y )$
    
При этом указанное правило должно подчиняться четырем аксиомам:
1) $( x, y ) = ( y, x )$

2) $( \lambda x, y )=\lambda( x, y )$

3) $( x_1 + x_2, y )=( x_1, y)+( x_2, y )$

4) $( x, x )\geq0$, причем $( x, x )=0\Leftrightarrow x=0.$


Рассмотрим третий пункт представленной аксиоматики и перепишем его с учетом заданного скалярного произведения:
$( x_1 + x_2, y )=\left \|x_1+x_2\right \| \cdot \left \|y\right \|$
$( x_1, y )+( x_2, y) = \left \|x_1\right \| \cdot \left \|y\right \| + \left \|x_2\right \| \cdot \left \|y\right \|$
Однако в общем случае $\left \|x_1+x_2\right \| \neq \left \|x_1\right \| + \left \|x_2\right \|$. Например, если мы пользуемся $\ell_{2}$ нормой и рассматриваем векторы $x_1 = (4, 2, 4); x_2 = (0, 0, 1)$. Тогда $\left \|x_1+x_2\right \| = \sqrt{45}$, в то время как $\left \|x_1\right \| + \left \|x_2\right \| = 6 + 1 =7$. 
Таким образом линейное пространство, в котором скалярное произведение векторов задается произведением длин этих векторов, не является евклидовым

б) В таком случае $( x, y )_{new} = 3 ( x, y )_0=3\cdot\left \|x\right \| \cdot \left \|y\right \| \cdot cos\alpha,$ где $\alpha$ - угол между векторами. Согласно определению, линейное пространство над полем вещественных чисел называется _евклидовым пространством_, если на нем введено правило, согласно которому каждой упорядоченной паре векторов $x$ и $y$ поставлено в соответствие вещественное число, называемое _скалярным произведением_ и обозначаемое символом $( x, y )$.

При этом указанное правило должно подчиняться четырем аксиомам:
1) $( x, y )=( y, x )$

2) $( \lambda x, y )=\lambda( x, y )$

3) $( x_1 + x_2, y )=( x_1, y )+( x_2, y )$

4) $( x, x )\geq0$, причем $( x, x )=0\Leftrightarrow x=0.$

Проверим все пункты по порядку, пользуясь тем фактом, что скалярного произведения $( x, y )_0$ указанные условия соблюдаются:
1) $( x, y )_{new} = 3 ( x, y )_0 = 3( y, x)_0 = ( y, x )_{new}$ 

2) $( \lambda x, y )_{new}=3( \lambda x, y )_0 =\lambda\cdot3( x, y )_0=\lambda( x, y )_{new}$

3) $( x_1 + x_2, y )_{new}=3( x_1 + x_2, y )_0 = 3( x_1, y )_0+3( x_2, y )_0 = ( x_1, y )_{new} + ( x_2, y )_{new}$

4) $(x, x )_{new}\geq0$, причем $( x, x )_{new}=0\Leftrightarrow x=0.$

Все условия для скалярного произведения заданного таким образом соблюдаются, а это значит, что линейное пространство является евклидовым

__4.__ Какие из нижеперечисленных векторов образуют ортонормированный базис в линейном пространстве $\mathbb{R}^{3}$:<br>
а) $(1,0,0),(0,0,1);$ <br>
б) $(1/\sqrt{2},-1/\sqrt{2},0),(1/\sqrt{2},1/\sqrt{2},0), (0,0,1);$<br>
в) $(1/2, -1/2, 0), (0, 1/2, 1/2), (0,0,1);$<br>
г) $(1,0,0),(0,1,0),(0,0,1)?$ 

 #### Решение:
 а) Указынные векторы не образуют базиса в линейном пространстве $\mathbb{R}^{3}$, так как к примеру вектор (0, 4, 0) не может быть представлен в виде линейной комбинации этих векторов
 
 б) $e_1 = (1/\sqrt{2},-1/\sqrt{2},0), e_2 = (1/\sqrt{2},1/\sqrt{2},0), e_3 = (0,0,1)$
 
Посчитаем все скалярные произведения:  
$( e_1, e_1 ) = \frac{1}{2} + \frac{1}{2} = 1$

$( e_1, e_2 )= \frac{1}{2} - \frac{1}{2} = 0$

$( e_1, e_3 ) = 0$

$( e_2, e_2 ) = \frac{1}{2} + \frac{1}{2} = 1$

$( e_2, e_3 ) = 0$

$( e_3, e_3 ) = 1$
Это означает, что представленный базис является ортонормированным

в) $e_1 = (1/2, -1/2, 0), e_2 = (0, 1/2, 1/2), e_3 = (0,0,1)$

Посчитаем скалярные произведения:  

$( e_1, e_1 ) = \frac{1}{4} + \frac{1}{4} = 1/2 \neq 1$

Значит данный базис уже точно не будет ортонормированным

г) $e_1 = (1,0,0),e_2 = (0,1,0),e_3 = (0,0,1)$ 

Посчитаем все скалярные произведения:
$( e_1, e_1 ) = 1$

$( e_1, e_2) = 0$

$(e_1, e_3 ) = 0$

$( e_2, e_2 ) = 1$

$( e_2, e_3) = 0$

$( e_3, e_3 ) = 1$

Это означает, что представленный базис является ортонормированным
